# 교차검증
1. Hold-out
- 개념 : 단순하게 Train(훈련용) 데이터를 (train, valid)라는 이름의 2개의 데이터로 나누는 작업으로서, 보통 8:2 또는 7:3으로 나눔
- 목적 : 예측 성능을 가늠해보기 위하여, 80%의 데이터를 통해 학습하고, 20%의 데이터를 예측함
- 단점 : 데이터의 낭비 #20%의 데이터는 모델이 학습할 기회도 없이 예측만 하고 버려짐
2. K-Fold (교차검증)
- 개념 : valid 데이터를 겹치지 않게 나누어 N개의 데이터셋을 만들어내는 과정을 반복함
- 목적 : Hold-out의 단점을 극복하기 위해, 모든 데이터를 최소한 한 번씩 학습하도록 함
- 코드 예시
    ```
    from sklearn.model_selection import KFold #KFold 불러오기
    kf = KFold(n_splits = 5, shuffle = True, random_state = 0) # 훈련 데이터를 5조각으로 나누는 교차검증모델 생성

    ```
- 원리 : 
    1. K-Fold를 이용해서 Train과 Valid Data를 나눈다.
    2. Model을 이용해서 train 데이터를 학습한다.
    3. Model을 이용해서 valid 데이터를 예측해 성능을 확인한다.
    4. Model을 이용해서 test 데이터를 예측한다.
    5. n_splits를 5로 설정한다면, 5개의 결과값들에 대한 “최빈값”을 이용해 가장 등장할 가능성이 높은 결과값으로 결정한다.
